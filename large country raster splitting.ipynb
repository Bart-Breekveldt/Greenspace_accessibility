```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import numpy as np
import warnings
```

```python
lower_48_DC = ['Washington State','Oregon','California','Nevada','Arizona','Utah','Idaho','Montana','Wyoming','Colorado','New Mexico',
          'North Dakota','South Dakota','Nebraska','Kansas','Oklahoma','Texas','Louisiana','Arkansas','Mississippi','Alabama',
          'Missouri','Tennessee','Kentucky','Iowa','Minnesota','Wisconsin','Illinois','Indiana','Michigan','Ohio','West Virginia',
          'Virginia','North Carolina','South Carolina','Georgia','Florida','Delaware','District of Columbia','Maryland',
          'Pennsylvania','New Jersey','New York State','Connecticut','Rhode Island','Massachusetts','Vermont','New Hampshire','Maine']
```

```python
len(lower_48_DC)
```

```python
warnings.filterwarnings('ignore')
gdf = gpd.GeoDataFrame(crs = 4326)
for i in lower_48_DC:
    gdf = pd.concat([gdf, ox.geocoder.geocode_to_gdf(i+', United States')])
    #print(i)
gdf.head()
```

```python
lower_48_dc_dissolved = gdf.dissolve().buffer(0.05)
```

```python
lower_48_dc_dissolved.to_file('D:/Dumps/lower48DC.shp')
```
